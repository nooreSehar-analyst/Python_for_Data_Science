```python
#Question No 2: Currency Risk Management (Arithmetic Operators)Aapka account balance 15,000 PKR hai aur aapne $50 ki subscription deni hai. Exchange rate 1 USD = 278 PKR hai.
conversion_amount = 50 * 278
account_balance = 15000
remaining_balance = account_balance - conversion_amount
print(f"Remaining Balance: ${remaining_balance}")
```
