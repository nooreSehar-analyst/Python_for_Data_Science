For loop

```python
#Task 1: Basic Iteration and Range Control
#Problem Statement: Write a Python program using a for loop and the range() function to print numbers from 0 to 10 (inclusive).
#Goal: To understand how the range() function's stop parameter works.
for i in range(11):
  print(i)
```

```python
#Task 2: Building Patterns with Nested Loops
#Problem Statement: Create a right-angled triangle pattern using stars (*) where the first line has 1 star, the second has 2, and the third has 3.
#Goal: To practice nested loops and the use of the end parameter in the print() function.
for i in range(1,6):
  for j in range(i):
    print("*", end=" ")
  print()
```

```python
# Task 3: Using Step Parameter in Range
#Problem Statement: Write a program to print only ODD numbers (1, 3, 5, 7, 9) using a single for loop and the range() function.
for i in range(1,10,2):
  print(i)
```

```python
#Task 4: Iteration with Custom Step Values
#Problem Statement: Write a Python program using a for loop and the range() function to print numbers starting from 5 up to 50, jumping by 5 steps each time.
#Goal: To understand how the step parameter in range(start, stop, step) controls the increment of the loop.
for i in range(5,51,5):
  print('step item:',i)
```

```python
#Task 5: Accessing List Elements using For Loop
#Problem Statement: Create a list named savings containing four different amounts. Write a for loop to iterate through this list and print each amount with a personalized message: "ye meri saving h: [amount]".
#Goal: To practice direct iteration over a List (Data Structure) instead of using range().
savings = [2000, 4500, 3100, 5000]
for i in savings:
  print('ye meri saving h:',i)
```

```python
#Task 6: Generating Multiples Using Step Parameter
#Problem Statement: Write a Python program using a single for loop and the range() function to print the multiples of 7, up to 70 (inclusive).
#Goal: To understand how to use the start, stop, and step arguments of the range() function effectively.
for i in range(7,71,7):
  print(i)
```

```python
#Task 7: Generating Multiplication Tables with String Formatting
#Problem Statement: Define a variable named table and assign it the number 7. Write a for loop to generate and print the multiplication table for this number from 1 to 10 in the standard format: "7 x i = result".
#Goal: To practice the integration of mathematical operations within f-strings inside a loop for clear output formatting.
table = 7
for i in range(1,11):
  print(f"{table} x {i} =",i*table)
```

```python
#Task 8: Dynamic Output using Variables and f-strings
#Problem Statement: Perform basic multiplication using variables and display the result using f-strings. Also, demonstrate a direct calculation inside a print() statement using the f"{}" syntax.
#Goal: To practice variable assignment and the ability to perform inline mathematical operations within strings for cleaner code.
num = 7
step= 1
result = 7*1
print(f"{num} multiply by {step} is equal to {result}")
```

```python
#Task 9: Automating Mathematical Tables
#Problem Statement: Use a for loop and the range() function to generate a multiplication table for the number 2. Ensure the output is formatted clearly using f-strings to show the full equation (e.g., 2 x 1 = 2).
#Goal: To solidify the concept of iterating through a range to automate repetitive mathematical tasks.
x = 2
for i in range (1,11):
  print(f" {x} x {i} = {x*i}")
```

```python
#Task 10: For more practuce
#Printing 10's Table using for loop
y= 10
for i in range(1,11):
  print(f"{y} x {i} = {y*i}")
```

```python
#Task 11: Generating Incremental Patterns using String Multiplication
#Problem Statement: Write a Python program using a for loop to print a right-angled triangle pattern. In each iteration, the number of stars (*) should increase by one, starting from 1 up to 10.
#Goal: To understand how string multiplication ("*" * i) can be used inside a loop to create visual patterns efficiently.
for i in range(1,11):
  print("*" * i)
```

```python
#Task 12: Creating Inverted Patterns with Reverse Range
#Problem Statement: Write a Python program to print an inverted triangle pattern starting with 11 stars and decreasing by one in each line until it reaches 2. Use a for loop with a negative step in the range() function.
#Goal: To practice using the step parameter (-1) in the range() function to iterate in reverse order for pattern generation.
for i in range(11,1,-1):
  print("*" *i)
```

```python
#Task 13: Multi-line Pattern Generation with Nested Loops
#Problem Statement: Create a right-angled triangle pattern using nested for loops. Use the end="" parameter in the print() function to stay on the same line for each row's stars.
#Goal: To understand how the inner loop's range can be dynamically controlled by the outer loop's variable (i).
for i in range(1, 6):          # Lines (1 se 5 tak)
    for j in range(i):         # Har line mein 'i' jitnay stars
        print("*", end="")     # end="" ka matlab hai agli line par mat jao
    print()                    # Jab andar wala loop khatam ho, to line change karo
```

Conditional Statement

```python
#Task : Create a program that takes the user's age as input and determines if they are eligible to vote (18 or older) or "too young to vote."
age = int(input('enter your age:'))
if age >= 18:
  print('you can vote')
else:
  print('too young to vote')
```

```python
#Task: Write a script to check whether a user-inputted number is Positive or Negative using an if-else block.
num = int(input('enter a number:'))
if num > 0:
  print('positive')
else:
  print('negative')
```

```python
#Task: Implement a billing system where a 10% discount is applied if the total amount exceeds 5000; otherwise, the user pays the full price.
bill_amount = int(input('enter a bill_amount:'))
if bill_amount >5000:
  print('you get a 10% discount!')
else:
  print('full price')
```
