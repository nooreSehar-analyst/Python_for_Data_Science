```python
# Task 22: make two set and find union, intersection and difference
a = {1,2,3,4,5}
b = {3,4,5,6,7}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

```python
# Task 23: Take a 5 number as input from the user and store them in a list. then print maximum number, minimum number, sum of all numbers
numbers = []
for i in range(5):
  num = int(input('enter number: '))
  numbers.append(num)
print(numbers)
print(max(numbers))
print(min(numbers))
print(sum(numbers))

```
